```python
# Bài 1:

def gcd(m, n):
    d = min(m, n) 
    while m % d != 0 or n % d != 0: 
        d -= 1 
    return d  


def frac_reduced(m, n):
    ucln = gcd(m, n)  
    return (m // ucln, n // ucln)  

m = int(input('Nhập m ='))
n = int(input('Nhập n ='))      
print("Ước số chung lớn nhất:", gcd(m, n))
print("Phân số rút gọn:", frac_reduced(m, n))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array(experience).reshape(-1, 1)  # Chuyển thành cột vector
y = np.array(salary)

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(x, y)

# Hệ số hồi quy
b0 = model.intercept_  
b1 = model.coef_[0]    

plt.scatter(x, y, color='red', label='Thực tế')  
plt.plot(x, model.predict(x), color='green', linewidth=2, label='Đường hồi quy')

plt.xlabel('Số năm kinh nghiệm')
plt.ylabel('Mức lương')
plt.title('Hồi quy tuyến tính: Lương vs. Số năm kinh nghiệm')
plt.legend()
plt.show()

print(f"Phương trình hồi quy: y = {b0:.2f} + {b1:.2f} * x")
```

```python
# Bài 4:
class Person:
    def __init__(self, name, age, gender, weight, height):
        self.name = name
        self.age = age
        self.gender = gender
        self.weight = weight
        self.height = height

    def info(self):
        return f"{self.name}, {self.age} tuổi, {self.gender}"

    def bmi(self):
        return self.weight / self.height ** 2
ds = [
    Person("Hà", 25, "Nữ", 55, 1.65),
    Person("Nam", 30, "Nam", 78, 1.75),
    Person("Hoà", 22, "Nữ", 95, 1.80),
    Person("kaka", 28, "Nam", 65, 1.60)
]

max_bmi = ds[0]
for person in ds[1:]:
    if person.bmi() > max_bmi.bmi():
        max_bmi = person

print(f"Người có BMI cao nhất: {max_bmi.name}, BMI = {max_bmi.bmi():.2f}")
```
