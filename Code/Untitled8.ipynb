```python
df = pd.read_excel('US Vehicle Traffic Fatalities.xlsx')
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Dữ Liệu 
x= df.Vehicles
y= df.Fatalities
# Nghiệm bình phương trối thiểu
A=np. vstack([np.ones(len(x)),x]).T
bo,b1=np.linalg.lstsq(A,y,rcond=None) [0]
print ("y=%f %f x"%(bo,b1))
```

```python
plt.scatter(x, y, color='blue', label='Dữ liệu gốc')
plt.plot(x, bo + b1 * x, color='red', label='Hàm hồi quy')
plt.xlabel('Vehicles')
plt.ylabel('Fatalities')
plt.title('Hồi quy tuyến tính giữa Vehicles và Fatalities')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
 
def f(x):
    return x**3 - 10*x**2 + 5

def dao_ham_f(x):
    return 3*x**2 - 20*x

x_0 = 0.5
sai_so = 1e-10

# Danh sách lưu kết quả
bang_ket_qua = []
lan_lap = 0

while True:
    gia_tri_f = f(x_0)
    dao_ham = dao_ham_f(x_0)

    if dao_ham == 0:
        break


    x = x_0 - gia_tri_f / dao_ham
    do_sai_lech = abs(x - x_0)


    bang_ket_qua.append([lan_lap, x_0, gia_tri_f, dao_ham, do_sai_lech])

    if do_sai_lech < sai_so:
        break
        
    x_0 = x
    lan_lap += 1
    
bang = pd.DataFrame(
    bang_ket_qua,
    columns=['Lần lặp', 'x', 'f(x)', "f'(x)", 'Sai lệch']
)

# Hiển thị bảng
print(bang)
```
